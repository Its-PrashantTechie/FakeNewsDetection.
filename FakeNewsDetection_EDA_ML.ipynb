```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set(style="ticks", color_codes=True)

# Combine True and Fake datasets (if not already done)
df_T_news = pd.read_csv("True.csv.zip")
df_F_news = pd.read_csv("Fake.csv.zip")
df_T_news['label'] = 'True'
df_F_news['label'] = 'Fake'
df = pd.concat([df_T_news, df_F_news], ignore_index=True)

# Plot stacked countplot (Fake vs True per subject)
fig_dims = (20, 6)
fig, ax = plt.subplots(figsize=fig_dims)

pd.crosstab(df['subject'], df['label']).plot(kind='bar', stacked=True, ax=ax, colormap='coolwarm')
plt.xticks(rotation=45)
plt.title('Fake vs True News per Subject')
plt.ylabel('Number of Articles')
plt.show()
```

```python
df_F_news['label'] = 0
df_T_news['label'] = 1
```

```python
import pandas as pd

df = pd.concat([df_F_news, df_T_news], ignore_index=True)
print(df.shape)
print(df['label'].value_counts())
```

```python
X = df['text']   # News content
y = df['label']  # 0 = Fake, 1 = True
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Train
model = PassiveAggressiveClassifier(max_iter=50)
model.fit(X_train_tfidf, y_train)

# Predict
y_pred = model.predict(X_test_tfidf)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
feature_names = vectorizer.get_feature_names_out()
coef = model.coef_[0]

top_fake_words = sorted(zip(coef, feature_names))[:20]   # most negative weights
top_true_words = sorted(zip(coef, feature_names), reverse=True)[:20]  # most positive weights

print("Top words indicating Fake news:", [w for _, w in top_fake_words])
print("Top words indicating True news:", [w for _, w in top_true_words])
```

```python
import joblib

joblib.dump(model, "fake_news_model.pkl")
joblib.dump(vectorizer, "tfidf_vectorizer.pkl")
```

```python
def predict_news_with_confidence(text, vectorizer, model):
    """
    Predict if a news article is Fake or True and show confidence.

    Args:
        text (str or list): Single news article (str) or multiple articles (list of str)
        vectorizer: Trained TfidfVectorizer
        model: Trained classifier (PassiveAggressiveClassifier, etc.)

    Returns:
        Prints prediction and confidence for each article
    """
    # Ensure input is a list
    if isinstance(text, str):
        text = [text]

    # Transform text
    vect_text = vectorizer.transform(text)

    # Predict
    pred = model.predict(vect_text)
    confidence = model.decision_function(vect_text)  # distance from hyperplane

    for i, article in enumerate(text):
        label = "True News ✅" if pred[i] == 1 else "Fake News ❌"
        conf_score = confidence[i]
        print(f"Article: {article}\nPrediction: {label} | Confidence: {conf_score:.2f}\n")

# Example usage:
new_articles = [
    " President Murmu to begin four-day Kerala visit from Tuesday ·",
    "Aliens landed in New York! People are freaking out."
]

predict_news_with_confidence(new_articles, vectorizer, model)
```
